### Ball_plot.py

```python
import numpy as np
import matplotlib.pyplot as plt

v0 = 5
g = 9.81

# The linspace Function: np.linspace[a,b] n : mean create 1001 coordinates between 0,1 
t = np.linspace(0,1,1001)

y = v0*t-0.5*g*t**2

plt.plot(t,y)      # plots all y coordinates vs. all t coordinates
plt.xlabel('t(s)') # places the text t (s) on x-axis
plt.ylabel('y(m)') # place the text y (m) on y-axis
plt.show()         # Display the figure 
```

### Assume we want to plot f(t) = t <sup>2</sup> and g(t) = e<sup>t</sup> in the same plot for t on the interval [-2, 2]. The following script will accomplish this task:

```python
import numpy as np 
import matplotlib.pyplot as plt

# choose 100 points in time interval
t = np.linspace(-2,2,100)

f_value = t**2 
g_value = np.exp(t)

plt.plot(t, f_value, 'r', t, g_value, 'b--') # គូសក្រាហ្វិកអនុគមន៍ f ពណ៌ក្រហម និង g ពណ៌ខៀវដិតៗ
plt.xlabel('t') # អ័ក្ស ដេក 
plt.ylabel('f and g') # អ័ក្សឈរ
plt.legend(['t**2','e**t']) # ក្នុងប្រអប់ប្រាប់ឈ្មោះអនុគមន៍នីមួយៗ
plt.title('Plotting of two functions (t**2 and e**t)') # ចំណងជើងបស់ប្រាប
plt.grid('on') # បំណែងចែកចន្លោះ 
plt.axis([-3,3,-1,10]) # ចន្លោះ អ័ក្សដេក និង អ័ក្សឈរត្រូវយកពីណាទីណា
plt.show()
```
